```python
wdi_indicators = ['NY.GDP.PCAP.PP.KD', 'NY.GDP.PCAP.KD', 'SL.GDP.PCAP.EM.KD', 
                  'SP.POP.GROW', 'SP.POP.TOTL', 'SP.DYN.WFRT', 'SP.DYN.TFRT.IN',
                  'NY.GDP.DEFL.KD.ZG', 'NY.GDP.MKTP.KD.ZG']
```

```python
wdi = wb.download(indicator=wdi_indicators, country=['RU','UA'], start=1950, end=2021)
wdi = wdi.reset_index()
wdi['year'] = wdi.year.astype(int)
wdi['gdp_pc'] = wdi['NY.GDP.PCAP.PP.KD']
wdi['ln_gdp_pc'] = wdi['NY.GDP.PCAP.PP.KD'].apply(np.log)
wdi['ln_pop'] = wdi['SP.POP.TOTL'].apply(np.log)
wdi.head()
```

```python
wdi['gdp_gr'] = wdi[ 'NY.GDP.MKTP.KD.ZG']
wdi['inflation'] = wdi['NY.GDP.DEFL.KD.ZG']
```

```python
wdi.tail()
```

```python
fig, ax = plt.subplots()
sns.lineplot(x='year', y='gdp_gr', data=wdi, ax=ax, 
                    hue='country',
                    alpha=1, 
                    style='country', 
                )
ax.set_xlabel('Year')
ax.set_ylabel('GDP Growth Rate')
ax.axvline(2014, c='r', linestyle=':')
ax.axvline(2016, c='r', linestyle=':')
ax.axvline(2022, c='r', linestyle=':')
```

```python
fig, ax = plt.subplots()
sns.lineplot(x='year', y='inflation', data=wdi, ax=ax, 
                    hue='country',
                    alpha=1, 
                    style='country', 
                )
ax.set_xlabel('Year')
ax.set_ylabel('Inflation Rate')
ax.axvline(2014, c='r', linestyle=':')
ax.axvline(2016, c='r', linestyle=':')
ax.axvline(2022, c='r', linestyle=':')
```
